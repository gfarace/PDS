### Análisis de ruidos cardíacos <br>

Los ruidos cardíacos son sonidos breves y transitorios producidos por la apertura y el cierre de las válvulas. Se obtienen mediante la auscultación del paciente utilizando un estetoscopio.
Los ruidos cardíacos se componen principalmente de dos sonidos: S1 y S2. El primer sonido (S1) está relacionado con el cierre de las válvulas auriculoventriculares, mientras que el segundo sonido (S2) señala el final de la sístole y el comienzo de la diástole y se puede escuchar en el momento del cierre de las válvulas aórticas y pulmonar [1]. El conjunto de estos dos sonidos corresponden a los ruidos normales de un corazón sano.

<img src="ruidos.png" width=500 />

Además de S1 y S2, existen también otros sonidos (S3 y S4) que denotan patologías cardíacas.

El rango de frecuencia del ruido cardíaco es generalmente de 5 a 600Hz. Por encima de 120Hz se considera como sonido de alta frecuencia; de 80Hz a 120Hz como sonido de frecuencia media; por debajo de 80Hz como sonido de baja frecuencia.
Para los ruidos S1 y S2 su frecuencia característica se encuentra entre 50 a 100Hz mientras que los ruidos S3 y S4 de 10 a 50Hz [2].

El análisis de los ruidos cardíacos se llevará a cabo de la siguiente forma: 

<b>Pre-procesamiento<br>
1) Downsampling: se lleva la señal a 2kHz <br>
2) Filtro pasabanda: Chebyshev entre las frecuencias de interés, de 30 a 195Hz <br>
3) Normalización <br>
4) Refiltrado: Butter pasa bajos para eliminación de ruido <br>
    
<b>Segmentación<br>
1) Detección de envolvente: aplicando energía de Shannon <br>
2) Identificación de picos <br>
3) Eliminación de picos <br>
4) Localización de S1 y S2 <br>


#### Referencias
